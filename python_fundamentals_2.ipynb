```python
#tuples are same as list but we cannot make changes to the items in the tuples
my_tuple=(1,2,3,4)
#2d tuple i.e. tuple inside tuple
hey=(1,2,3,(True,'ram'))
hey[-1][1]


#tuple unpacking means assigning values to tuple to the variables
a,b,c=(2,3,4)
print(a)
a,b,*others=(2,3,4,5,56,33)
print(others)

#we can do zipping in tuples
```

```python
#sets (it consists unique values and we can do union,intersection and all)
#sets are mutable but they only can contain immutable values inside them like set inside set is invalid,list inside set is invlid but can has tuple as it is immutable.
#we cannot do indexing and access values with indexing in sets
set={(1,2,3),5,6,7}
a=list(set)
set.add('hello')
print(set)
```

```python
#set operations
set1={1,2,3,4,5}
set2={2,3,6,7,8}
set1.union(set2)
print(set1.intersection(set2))
print(set1&set2)
print(set1|set2)
print(set1-set2)
print(set1^set2) #it is symmetric difference means elements except common

#we can do "in" and "not in" commands in set too.
#we can use len, min, max, sort all in sets too.
```

```python
#frozen set are sets which are immutable
set=frozenset([1,2,3])
set


#we can also do set comprehension here
set={i**2 for i in range(1,11) if i%2==0}
set
```
